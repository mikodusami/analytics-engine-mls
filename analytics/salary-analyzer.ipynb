```python
# this function represents the cleanse of each group in missing_values_df
def cleanse_null_values(data):
    # main_df should be the main dataframe
    _missing_values_df = df[df.isna().any(axis=1)]
    print(f"there are {_missing_values_df.shape[0]} rows containing null values")
    _years_split_dict = {year: group for year, group in missing_values_df.groupby('year')}
    print(f"there are null values found within each of these years: {_years_split_dict.keys()}")
    # --- YEAR 2007-2009 ---
    _current_years_analyzing = [2007, 2008, 2009]
    for year in _current_years_analyzing:
        current_year_df_missing = _years_split_dict[year].isna().any(axis=1)
        data.drop(index=current_year_df_missing.index, inplace=True)
    
  
    # --- YEAR 2010 ---
    # -- we need to drop all rows where first_name and salary is na()
    current_year_analyzing = 2010
    current_year_df = _years_split_dict[current_year_analyzing]
    mask = (current_year_df["first_name"].isna()) & (current_year_df['base_salary'].astype(str).str.strip() == str(current_year_analyzing))
    to_drop = current_year_df.loc[mask]
    data.drop(index=to_drop.index, inplace=True)
    
    
    # --- YEAR 2011 ---
    current_year_analyzing = 2011
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2012 ---
    current_year_analyzing = 2012
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2013 ---
    current_year_analyzing = 2013
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2014 ---
    current_year_analyzing = 2014
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2015 ---
    current_year_analyzing = 2015
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2016 ---
    current_year_analyzing = 2016
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2017 ---
    current_year_analyzing = 2017
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2018 ---
    current_year_analyzing = 2018
    current_year_df = _years_split_dict[current_year_analyzing]

    # --- YEAR 2019 ---
    # current_year_analyzing = 2019
    # current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2020 ---
    current_year_analyzing = 2020
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2021 ---
    current_year_analyzing = 2021
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2022 ---
    current_year_analyzing = 2022
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2023 ---
    current_year_analyzing = 2023
    current_year_df = _years_split_dict[current_year_analyzing]
    
    # --- YEAR 2024 ---
    current_year_analyzing = 2024
    current_year_df = _years_split_dict[current_year_analyzing]

    data['club'] = data['club'].fillna('N/A')
    
    return data
```
